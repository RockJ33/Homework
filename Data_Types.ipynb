```python
#12.) Create a set of about 3 elements
set1 = {"car", "truck", "train"} 

#Add a list of fruits to the above set and print the result
fruitset = {"apple", "orange", "grape"}
set1.update(fruitset)
print(set1)

#Add a car element to your set
# carset = {"Camaro", "Mustang", "Porsche", "Ferrari"}
# set1.update(carset)
# print(set1) #or

fruitset.add("car element")
print(fruitset)

# creat a 2nd set with a few odd items
set2 = {"store", "road", "computer", "fox", "oven"}

# Save the union of set1 and set2 to a 3rd set
set3 = set1.union(set2)
print(set3)

#Pop and element from tehe second set and print remainder
set2.pop()
print(set2)

#Clear the first set and print the result
set1.clear()
print("This set is cleared", set1)

# Discard an element, and remove another element from the 3rd set
set3.discard("store")
set3.remove("grape")

# Print the remainder of the 3rd set
print(set3)
```
